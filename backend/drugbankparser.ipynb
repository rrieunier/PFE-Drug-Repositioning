```python
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
root = ET.parse('./full database.xml').getroot()
ns = '{http://www.drugbank.ca}'
```

```python
root.findall(f"{ns}drug")
```

```python
# [(tag.find(f'{ns}name').text) for tag in root.getchildren()]

productlist = {}
for tag in root.findall(f"{ns}drug"):
    productlist[tag.find(f"{ns}name").text] = (set([product.find(f"{ns}name").text for product in tag.find(f'{ns}products').findall(f'{ns}product')]))
    
productlist

# [product for product in tag.find(f'{ns}products').getchildren() for tag in root.getchildren()][0]
```

```python
element = root.findall(f"{ns}drug")[0]
element.find(f"{ns}name").text
```

```python
drugs = []
for drug in root.findall(f"{ns}drug"):
    for product in drug.find(f'{ns}products').findall(f'{ns}product'):
        drugs.append({"Name": drug.find(f"{ns}name").text, "Labeller": drug.find(f"{ns}description").text, "Countries": product.find(f"{ns}country").text, "Drug": product.find(f"{ns}name").text})

pd.DataFrame(drugs).to_csv("drugbank.csv", sep=";")
```
